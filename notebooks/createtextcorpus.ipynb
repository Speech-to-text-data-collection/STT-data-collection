```python
import glob
import json
```

```python
with open('../data/preprocessed/data_0.txt', encoding='utf-8') as textfile:
    lines = textfile.readlines()
```

```python
unicode_amharic_range = (4608, 4955)
def remove_chars(sentence:str) -> str:
    new_sentence = ''
    for char in sentence:
        if(ord(char) >= unicode_amharic_range[0] and ord(char) <= unicode_amharic_range[1] or ord(char) == 32):
            new_sentence += char

    new_sentence.replace("  "," ")

    return new_sentence
```

```python
cleaned_text = list(map(remove_chars, lines))
```

```python
example_text_corpus = {}
for i in range(len(cleaned_text)):
    example_text_corpus[i] = cleaned_text[i]
```

```python
with open('../data/example_text_corpus.json', mode='w' ,encoding='utf-8') as obj:
    json.dump(example_text_corpus, obj, ensure_ascii=False, indent=4)
```
